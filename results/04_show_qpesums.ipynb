This notebook generate the 6 snapshot of 1 hour QPESUMS data.

```python
# Utility to plot a 6*275*162 numpy array
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from inspect import cleandoc as dedent

import os
from mpl_toolkits.basemap import Basemap

lat0 = 21.8875
lat1 = 25.3125
lon0 = 120.0
lon1 = 122.0125

lats = np.arange(21.8875, 25.3125, 0.0125)
lons = np.arange(120.0, 122.0125, 0.0125)

def plot_dbz_6maps(data):
    import matplotlib.pyplot as plt
    # Get data dimensions
    nl, ny, nx = data.shape
    print('Data dimensions: ' + str(nl) + ' layers of ' + str(ny) + ' x ' + str(nx))
    # Making plot
    for i in range(nl):
        ax = plt.subplot(2, nl/2, i+1)
        # create basemap
        m = Basemap(llcrnrlon=lon0, urcrnrlon=lon1, llcrnrlat=lat0, urcrnrlat=lat1, resolution='l')
        # draw coastlines on map.
        m.drawcoastlines()
        m.imshow(((data[i, :,:])), alpha=0.99, cmap='Greys', vmin=0., vmax=60)
        m.colorbar()
        #ax.get_xaxis().set_visible(False)
        #ax.get_yaxis().set_visible(False)
    #plt.tight_layout()
    plt.show()
    return(0)


# Full Dataset (34369 records)
d1 = np.load('../data/dbz_2014061316_LST/2014061412.npy', allow_pickle=True)
d2 = np.load('../data/dbz_2014061316/2014061412.npy', allow_pickle=True)
#plot_dbz_6maps(d1)
plot_dbz_6maps(d2)
```
